```python
x=df.iloc[:,0:13]
Y=df['Heart Disease']
```

```python
Label_Encoder=LabelEncoder()
y=Label_Encoder.fit_transform(Y)
y
```

```python
train_tar=pd.DataFrame(y)
train_tar
```

```python
stand_scaler=StandardScaler()
X=stand_scaler.fit_transform(x)
X
```

```python
train_fur=pd.DataFrame(X)
train_fur
```

```python
df_crr=train_fur.corr()
df_crr
```

```python
plt.figsize=(20,50)
sns.heatmap(df_crr,annot=True)
plt.title("Heat Map")
plt.tight_layout()
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_fur, train_tar, test_size=0.33, random_state=42)
```

```python
lr_model=LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pre=lr_model.predict(X_train)
lr_pred=lr_model.predict(X_test)
lr_accur=accuracy_score(y_test,lr_pred)



```

```python
l_t_a=accuracy_score(y_train,lr_pre)
l_t_f=f1_score(y_train,lr_pre)
l_te_f=f1_score(y_test,lr_pred)
print("accuracy : %s" % l_t_a)
print("f1-score : %s" % l_t_f)
print(lr_accur)
print(l_te_f)
```

```python
from sklearn.metrics import confusion_matrix , ConfusionMatrixDisplay
cm = confusion_matrix(y_test, lr_pred, labels = lr_model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = lr_model.classes_)

disp.plot(cmap = 'Blues')
disp.figure_.suptitle('Logistic Regression Confusion Matrix')
# disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
# disp.figure_.suptitle("Confusion Matrix")
# print(f"Confusion matrix:\n{disp.confusion_matrix}")
plt.savefig('lr_cm.png',format = 'png')
plt.show()
```

```python
svc_model=SVC()
svc_model.fit(X_train,y_train)
svc_pre=svc_model.predict(X_train)
svc_pred=svc_model.predict(X_test)
svc_accur=accuracy_score(y_test,svc_pred)
```

```python
s_t_a=accuracy_score(y_train,svc_pre)
s_t_f=f1_score(y_train,svc_pre)
s_te_f=f1_score(y_test,svc_pred)
print("accuracy : %s" % s_t_a)
print("f1-score : %s" % s_t_f)
print(svc_accur)
print(s_te_f)
```
